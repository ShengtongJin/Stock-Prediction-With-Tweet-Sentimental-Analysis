```python
import datetime
import pandas as pd
import collections

df = pd.read_csv('../data/AMD_Sentiment.csv')
date_list = df.date.tolist()
date_list = list(dict.fromkeys(date_list))


result=pd.DataFrame()

open_score = dict()
close_score = dict()

# next_date=[]
for i,r in df.iterrows():
    date=str(r.date)
    date1 = datetime.datetime.strptime(str(r.date),"%Y-%m-%d") #.strftime("%Y-%m-%d")
    delta = datetime.timedelta(days = 1)
    next_date = date1+delta

    time=datetime.datetime.strptime(str(r.time),'%H:%M:%S')
    time_open=datetime.datetime.strptime('09:00:00','%H:%M:%S')
    time_close=datetime.datetime.strptime('16:00:00','%H:%M:%S')
    
    #before open time
    #opening price of current day
    if time<=time_open:
        #continue
        if date not in open_score:
            open_score[date] = [r.sentiment]
        else:
            open_score[date].append(r.sentiment)
    elif time <= time_close: 
        if date not in close_score:
            close_score[date] = [r.sentiment]
        else:
            close_score[date].append(r.sentiment)
    else:
        if next_date not in open_score:
            open_score[next_date.strftime("%Y-%m-%d")] = [r.sentiment]
        else:
            open_score[next_date.strftime("%Y-%m-%d")].append(r.sentiment)



for key,value in open_score.items():
    open_score[key] = sum(value)/len(value)


for key,value in close_score.items():
    close_score[key] = sum(value)/len(value)

df1 = pd.DataFrame(columns=["open_score", "close_score", "date"])

df1.date = date_list


for i,r in df1.iterrows():
    if r.date in open_score:
        df1.at[i,'open_score'] = open_score[r.date]
    else:
        df1.at[i,'open_score']= 0

    if r.date in close_score:
        df1.at[i,'close_score'] = close_score[r.date]
    else:
        df1.at[i,'close_score'] = 0


df1.to_csv('../data/AMD_score.csv',sep=',',encoding='utf-8')


df1.tail(10)

```
