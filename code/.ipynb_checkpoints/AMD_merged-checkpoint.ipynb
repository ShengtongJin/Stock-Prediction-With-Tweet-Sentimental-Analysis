```python
import pandas as pd
import copy

#read file

filenames=['../data/AMD_score.csv','../data/AMD.csv']

df1=pd.read_csv(filenames[0])
df2=pd.read_csv(filenames[1])

df3 = pd.merge(df1,df2, how = "inner", left_on = ["date"], right_on = ["Date"])
df3.drop(df3.columns[0],axis=1,inplace=True)

df3.to_csv('../data/merged_AMD.csv', sep=',', encoding='utf-8')

df3.tail(10)
```
